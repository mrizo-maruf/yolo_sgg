```python
import os
import json
import math
import glob
from dataclasses import dataclass
from typing import Dict, List, Tuple

import numpy as np
import matplotlib.pyplot as plt


ROOT = "/home/yehia/rizo/THUD_Robot/Synthetic_Scenes"
OUT_DIR = os.path.join("/home/yehia/rizo/code/yolo_sgg", "scores", "thud_synth")

# The 5 metrics (pentagon radar)
RADAR_METRICS = ["MOTA", "MOTP", "ID_consistency", "T_SR", "T_mIoU"]
```

```python
@dataclass
class MetricsRow:
    json_path: str
    scene: str
    rel_label: str  # label to show in legend
    MOTA: float
    MOTP: float
    ID_consistency: float
    T_SR: float
    T_mIoU: float
    MOTA_FN_ratio: float
    MOTA_FP_ratio: float
    MOTA_IDSW_ratio: float


def safe_float(d: dict, key: str, default: float = float("nan")) -> float:
    v = d.get(key, default)
    try:
        return float(v)
    except Exception:
        return default


def discover_jsons(root: str) -> List[str]:
    # You can tighten this pattern if needed:
    # pattern = os.path.join(root, "**", "*_metrics.json")
    pattern = os.path.join(root, "**", "*metrics.json")
    return sorted(glob.glob(pattern, recursive=True))


def scene_from_path(root: str, p: str) -> str:
    rel = os.path.relpath(p, root)
    parts = rel.split(os.sep)
    return parts[0] if parts else "UNKNOWN_SCENE"


def clamp01(x: float) -> float:
    if math.isnan(x):
        return x
    return max(0.0, min(1.0, x))


def load_rows(paths: List[str], root: str) -> List[MetricsRow]:
    rows: List[MetricsRow] = []
    for p in paths:
        try:
            with open(p, "r") as f:
                d = json.load(f)
        except Exception as e:
            print(f"[WARN] Failed to load {p}: {e}")
            continue

        scene = scene_from_path(root, p)
        rel = os.path.relpath(p, root)

        row = MetricsRow(
            json_path=p,
            scene=scene,
            rel_label=rel,
            MOTA=clamp01(safe_float(d, "MOTA")),
            MOTP=clamp01(safe_float(d, "MOTP")),
            ID_consistency=clamp01(safe_float(d, "ID_consistency")),
            T_SR=clamp01(safe_float(d, "T_SR")),
            T_mIoU=clamp01(safe_float(d, "T_mIoU")),
            MOTA_FN_ratio=clamp01(safe_float(d, "MOTA_FN_ratio")),
            MOTA_FP_ratio=clamp01(safe_float(d, "MOTA_FP_ratio")),
            MOTA_IDSW_ratio=clamp01(safe_float(d, "MOTA_IDSW_ratio")),
        )
        rows.append(row)
    return rows


def radar_plot_for_scene(scene: str, scene_rows: List[MetricsRow], out_dir: str) -> str:
    # Radar setup
    labels = RADAR_METRICS
    n = len(labels)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False).tolist()
    angles += angles[:1]  # close polygon

    fig = plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    # Put first axis at the top and go clockwise
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    # Axis labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_ylim(0.0, 1.0)

    # Plot each capture/json as one polygon
    for r in scene_rows:
        vals = [getattr(r, m) for m in labels]
        vals = [0.0 if (v is None or (isinstance(v, float) and math.isnan(v))) else float(v) for v in vals]
        vals += vals[:1]
        ax.plot(angles, vals, linewidth=2, label=r.rel_label)
        ax.fill(angles, vals, alpha=0.08)

    ax.set_title(f"Radar Metrics — Scene: {scene}", pad=18)

    # Legend outside (right)
    ax.legend(loc="center left", bbox_to_anchor=(1.15, 0.5), fontsize=8, frameon=True)

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, f"radar_{scene}.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path


def mota_breakdown_and_waterfall(rows: List[MetricsRow], out_dir: str) -> str:
    # Prepare labels
    labels = [r.rel_label for r in rows]
    x = np.arange(len(rows))

    fn = np.array([r.MOTA_FN_ratio for r in rows], dtype=float)
    fp = np.array([r.MOTA_FP_ratio for r in rows], dtype=float)
    ids = np.array([r.MOTA_IDSW_ratio for r in rows], dtype=float)
    mota = np.array([r.MOTA for r in rows], dtype=float)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

    # --- Panel 1: Stacked bar breakdown ---
    ax1.bar(x, fn, label="FN ratio")
    ax1.bar(x, fp, bottom=fn, label="FP ratio")
    ax1.bar(x, ids, bottom=fn + fp, label="IDSW ratio")
    ax1.set_title("MOTA Penalty Breakdown (Stacked)")
    ax1.set_ylabel("Penalty ratio (0..1)")
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels, rotation=90, fontsize=7)
    ax1.set_ylim(0.0, 1.0)
    ax1.legend()

    # --- Panel 2: Waterfall per JSON (grouped) ---
    # Waterfall steps: start=1.0, -FN, -FP, -IDSW, end=MOTA
    step_names = ["Start(1.0)", "-FN", "-FP", "-IDSW", "MOTA"]
    steps = 5

    group_gap = 1.5
    bar_w = 0.8

    for i, r in enumerate(rows):
        base = 1.0
        deltas = [0.0, -r.MOTA_FN_ratio, -r.MOTA_FP_ratio, -r.MOTA_IDSW_ratio, 0.0]
        # cumulative after each step
        cum = [base]
        cum.append(cum[-1] + deltas[1])
        cum.append(cum[-1] + deltas[2])
        cum.append(cum[-1] + deltas[3])
        cum.append(r.MOTA)

        # positions for this group
        start_pos = i * (steps + group_gap)
        pos = np.arange(steps) + start_pos

        # bars: first bar is base level; middle bars are deltas; last bar is final value
        # We'll draw each bar from previous cumulative level.
        prev = base
        # Start bar
        ax2.bar(pos[0], base, width=bar_w, label="Start" if i == 0 else None)
        # Delta bars
        for k in [1, 2, 3]:
            new_val = cum[k]
            height = deltas[k]
            # bar starts at prev if negative, else at prev (not expected here)
            ax2.bar(
                pos[k],
                height,
                bottom=prev,
                width=bar_w,
                label=("Penalty" if (i == 0 and k == 1) else None),
            )
            prev = new_val
        # Final bar (MOTA)
        ax2.bar(pos[4], r.MOTA, width=bar_w, label="MOTA" if i == 0 else None)

        # Group label under the middle
        mid = start_pos + (steps - 1) / 2
        ax2.text(mid, -0.10, os.path.basename(os.path.dirname(r.json_path)), ha="center", va="top", fontsize=7,
                 rotation=0, transform=ax2.get_xaxis_transform())

    ax2.set_title("MOTA Waterfall (All JSONs)")
    ax2.set_ylabel("Score / Cumulative")
    ax2.set_ylim(-0.1, 1.05)

    # x ticks: show step names repeated once (on first group), keep clean
    first_group = np.arange(steps)
    ax2.set_xticks(first_group)
    ax2.set_xticklabels(step_names, rotation=0, fontsize=9)

    ax2.legend()

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, "mota_breakdown_and_waterfall.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path
```

```python
os.makedirs(OUT_DIR, exist_ok=True)

paths = discover_jsons(ROOT)
# Save discovered paths
paths_txt = os.path.join(OUT_DIR, "discovered_metrics_json_paths.txt")
with open(paths_txt, "w") as f:
    for p in paths:
        f.write(p + "\n")
print(f"[OK] Discovered {len(paths)} json files.")
print(f"[OK] Saved list: {paths_txt}")

rows = load_rows(paths, ROOT)

# Group by scene
by_scene: Dict[str, List[MetricsRow]] = {}
for r in rows:
    by_scene.setdefault(r.scene, []).append(r)

# Radar plots per scene
radar_outs = []
for scene, scene_rows in sorted(by_scene.items(), key=lambda kv: kv[0]):
    out = radar_plot_for_scene(scene, scene_rows, OUT_DIR)
    radar_outs.append(out)
    print(f"[OK] Radar saved: {out}")

# Combined breakdown + waterfall for all jsons
combo = mota_breakdown_and_waterfall(rows, OUT_DIR)
print(f"[OK] Breakdown+waterfall saved: {combo}")

print(f"\nAll outputs saved under: {OUT_DIR}")
```

```python

from typing import Dict, List, Optional
# If True: annotate every JSON polygon in per-scene radar (can be unreadable if many jsons)
ANNOTATE_ALL_POLYGONS_PER_SCENE = False


@dataclass
class MetricsRow:
    json_path: str
    scene: str
    rel_label: str
    MOTA: float
    MOTP: float
    ID_consistency: float
    T_SR: float
    T_mIoU: float
    MOTA_FN_ratio: float
    MOTA_FP_ratio: float
    MOTA_IDSW_ratio: float


def safe_float(d: dict, key: str, default: float = float("nan")) -> float:
    v = d.get(key, default)
    try:
        return float(v)
    except Exception:
        return default


def clamp01(x: float) -> float:
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return float("nan")
    return max(0.0, min(1.0, float(x)))


def discover_jsons(root: str) -> List[str]:
    pattern = os.path.join(root, "**", "*metrics.json")
    return sorted(glob.glob(pattern, recursive=True))


def scene_from_path(root: str, p: str) -> str:
    rel = os.path.relpath(p, root)
    parts = rel.split(os.sep)
    return parts[0] if parts else "UNKNOWN_SCENE"


def load_rows(paths: List[str], root: str) -> List[MetricsRow]:
    rows: List[MetricsRow] = []
    for p in paths:
        try:
            with open(p, "r") as f:
                d = json.load(f)
        except Exception as e:
            print(f"[WARN] Failed to load {p}: {e}")
            continue

        scene = scene_from_path(root, p)
        rel = os.path.relpath(p, root)

        rows.append(
            MetricsRow(
                json_path=p,
                scene=scene,
                rel_label=rel,
                MOTA=clamp01(safe_float(d, "MOTA")),
                MOTP=clamp01(safe_float(d, "MOTP")),
                ID_consistency=clamp01(safe_float(d, "ID_consistency")),
                T_SR=clamp01(safe_float(d, "T_SR")),
                T_mIoU=clamp01(safe_float(d, "T_mIoU")),
                MOTA_FN_ratio=clamp01(safe_float(d, "MOTA_FN_ratio")),
                MOTA_FP_ratio=clamp01(safe_float(d, "MOTA_FP_ratio")),
                MOTA_IDSW_ratio=clamp01(safe_float(d, "MOTA_IDSW_ratio")),
            )
        )
    return rows


def radar_axes_setup(ax, metric_labels):
    n = len(metric_labels)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False).tolist()
    angles += angles[:1]

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metric_labels)
    ax.set_ylim(0.0, 1.0)
    return angles


def metric_vector_from_dict(d: Dict[str, float]) -> List[float]:
    vals = []
    for m in RADAR_METRICS:
        v = d.get(m, float("nan"))
        if v is None or (isinstance(v, float) and math.isnan(v)):
            v = 0.0
        vals.append(float(v))
    return vals


def metric_vector_from_row(r: MetricsRow) -> List[float]:
    vals = []
    for m in RADAR_METRICS:
        v = getattr(r, m)
        if v is None or (isinstance(v, float) and math.isnan(v)):
            v = 0.0
        vals.append(float(v))
    return vals


def annotate_radar_values(
    ax,
    angles: List[float],
    vals_closed: List[float],
    label: str,
    fmt: str = "{:.3f}",
    r_offset: float = 0.03,
    fontsize: int = 8,
):
    """
    Put numeric values near each vertex.
    angles includes closure; we annotate only first N vertices (exclude last).
    vals_closed includes closure.
    """
    n = len(RADAR_METRICS)
    for i in range(n):
        ang = angles[i]
        r = vals_closed[i]
        rr = min(1.0, r + r_offset)
        ax.text(
            ang,
            rr,
            fmt.format(r),
            ha="center",
            va="center",
            fontsize=fontsize,
        )


def compute_scene_averages(by_scene: Dict[str, List[MetricsRow]]) -> Dict[str, Dict[str, float]]:
    scene_avg: Dict[str, Dict[str, float]] = {}
    for scene, rows in by_scene.items():
        scene_avg[scene] = {}
        for m in RADAR_METRICS:
            arr = np.array([getattr(r, m) for r in rows], dtype=float)
            scene_avg[scene][m] = clamp01(float(np.nanmean(arr)) if np.isfinite(arr).any() else float("nan"))
    return scene_avg


def compute_global_average(rows: List[MetricsRow]) -> Dict[str, float]:
    g: Dict[str, float] = {}
    for m in RADAR_METRICS:
        arr = np.array([getattr(r, m) for r in rows], dtype=float)
        g[m] = clamp01(float(np.nanmean(arr)) if np.isfinite(arr).any() else float("nan"))
    return g


def radar_plot_per_scene(scene: str, scene_rows: List[MetricsRow], scene_avg: Optional[Dict[str, float]], out_dir: str) -> str:
    fig = plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)
    angles = radar_axes_setup(ax, RADAR_METRICS)

    for r in scene_rows:
        vals = metric_vector_from_row(r)
        vals_closed = vals + vals[:1]
        ax.plot(angles, vals_closed, linewidth=2, label=r.rel_label)
        ax.fill(angles, vals_closed, alpha=0.08)

        if ANNOTATE_ALL_POLYGONS_PER_SCENE:
            annotate_radar_values(ax, angles, vals_closed, label=r.rel_label, fontsize=7)

    # Annotate scene average polygon (recommended)
    if scene_avg is not None:
        vals = metric_vector_from_dict(scene_avg)
        vals_closed = vals + vals[:1]
        ax.plot(angles, vals_closed, linewidth=3, label=f"{scene} AVG")
        ax.fill(angles, vals_closed, alpha=0.12)
        annotate_radar_values(ax, angles, vals_closed, label=f"{scene} AVG", fontsize=9, r_offset=0.04)

    ax.set_title(f"Radar — Scene: {scene} (JSON polygons + scene AVG annotated)", pad=18)
    ax.legend(loc="center left", bbox_to_anchor=(1.15, 0.5), fontsize=8, frameon=True)

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, f"radar_per_json_{scene}.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path


def radar_plot_all_scenes(scene_avg: Dict[str, Dict[str, float]], out_dir: str) -> str:
    fig = plt.figure(figsize=(9, 9))
    ax = plt.subplot(111, polar=True)
    angles = radar_axes_setup(ax, RADAR_METRICS)

    for scene in sorted(scene_avg.keys()):
        vals = metric_vector_from_dict(scene_avg[scene])
        vals_closed = vals + vals[:1]
        ax.plot(angles, vals_closed, linewidth=2, label=scene)
        ax.fill(angles, vals_closed, alpha=0.08)
        annotate_radar_values(ax, angles, vals_closed, label=scene, fontsize=8, r_offset=0.03)

    ax.set_title("Radar — All Scenes (each polygon = scene avg, values annotated)", pad=18)
    ax.legend(loc="center left", bbox_to_anchor=(1.20, 0.5), fontsize=9, frameon=True)

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, "radar_scene_averages_all_scenes.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path


def radar_plot_global_average(global_avg: Dict[str, float], out_dir: str) -> str:
    fig = plt.figure(figsize=(7.5, 7.5))
    ax = plt.subplot(111, polar=True)
    angles = radar_axes_setup(ax, RADAR_METRICS)

    vals = metric_vector_from_dict(global_avg)
    vals_closed = vals + vals[:1]

    ax.plot(angles, vals_closed, linewidth=3, label="GLOBAL_AVG_ALL_JSONS")
    ax.fill(angles, vals_closed, alpha=0.12)
    annotate_radar_values(ax, angles, vals_closed, label="GLOBAL", fontsize=10, r_offset=0.05)

    ax.set_title("Radar — Global Average (ALL JSONs, values annotated)", pad=18)
    ax.legend(loc="upper right", bbox_to_anchor=(1.15, 1.10), frameon=True)

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, "radar_all_jsons_global_average.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path


def mota_breakdown_and_waterfall(rows: List[MetricsRow], out_dir: str) -> str:
    labels = [r.rel_label for r in rows]
    x = np.arange(len(rows))

    fn = np.array([r.MOTA_FN_ratio for r in rows], dtype=float)
    fp = np.array([r.MOTA_FP_ratio for r in rows], dtype=float)
    ids = np.array([r.MOTA_IDSW_ratio for r in rows], dtype=float)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

    ax1.bar(x, fn, label="FN ratio")
    ax1.bar(x, fp, bottom=fn, label="FP ratio")
    ax1.bar(x, ids, bottom=fn + fp, label="IDSW ratio")
    ax1.set_title("MOTA Penalty Breakdown (All JSONs)")
    ax1.set_ylabel("Penalty ratio (0..1)")
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels, rotation=90, fontsize=7)
    ax1.set_ylim(0.0, 1.0)
    ax1.legend()

    step_names = ["Start(1.0)", "-FN", "-FP", "-IDSW", "MOTA"]
    steps = 5
    group_gap = 1.5
    bar_w = 0.8

    for i, r in enumerate(rows):
        base = 1.0
        deltas = [0.0, -r.MOTA_FN_ratio, -r.MOTA_FP_ratio, -r.MOTA_IDSW_ratio, 0.0]

        cum = [base]
        cum.append(cum[-1] + deltas[1])
        cum.append(cum[-1] + deltas[2])
        cum.append(cum[-1] + deltas[3])
        cum.append(r.MOTA)

        start_pos = i * (steps + group_gap)
        pos = np.arange(steps) + start_pos

        prev = base
        ax2.bar(pos[0], base, width=bar_w, label="Start" if i == 0 else None)

        for k in [1, 2, 3]:
            height = deltas[k]
            ax2.bar(pos[k], height, bottom=prev, width=bar_w, label="Penalty" if (i == 0 and k == 1) else None)
            prev = cum[k]

        ax2.bar(pos[4], r.MOTA, width=bar_w, label="MOTA" if i == 0 else None)

        mid = start_pos + (steps - 1) / 2
        ax2.text(mid, -0.10, os.path.basename(r.json_path), ha="center", va="top", fontsize=6,
                 transform=ax2.get_xaxis_transform())

    ax2.set_title("MOTA Waterfall (All JSONs)")
    ax2.set_ylabel("Score / Cumulative")
    ax2.set_ylim(-0.1, 1.05)
    ax2.set_xticks(np.arange(steps))
    ax2.set_xticklabels(step_names, fontsize=9)
    ax2.legend()

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, "mota_breakdown_and_waterfall.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path

```

```python
def radar_plot_global_average(global_avg: Dict[str, float], out_dir: str) -> str:
    """NEW: single polygon radar for global average over all JSONs."""
    fig = plt.figure(figsize=(7.5, 7.5))
    ax = plt.subplot(111, polar=True)
    angles = radar_axes_setup(ax, RADAR_METRICS)

    vals = []
    for m in RADAR_METRICS:
        v = global_avg.get(m, float("nan"))
        if v is None or (isinstance(v, float) and math.isnan(v)):
            v = 0.0
        vals.append(float(v))
    vals += vals[:1]

    ax.plot(angles, vals, linewidth=3, label="GLOBAL_AVG_ALL_JSONS")
    ax.fill(angles, vals, alpha=0.12)

    ax.set_title("Radar — Global Average (ALL JSONs)", pad=18)
    ax.legend(loc="upper right", bbox_to_anchor=(1.15, 1.10), frameon=True)

    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, "radar_all_jsons_global_average.png")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return out_path
```

```python
def compute_global_average(rows: List[MetricsRow]) -> Dict[str, float]:
    """NEW: average across ALL json rows -> one polygon for the whole dataset."""
    g: Dict[str, float] = {}
    for m in RADAR_METRICS:
        arr = np.array([getattr(r, m) for r in rows], dtype=float)
        g[m] = clamp01(float(np.nanmean(arr)) if np.isfinite(arr).any() else float("nan"))
    return g
```

```python
import numpy as np
from collections import defaultdict

RADAR_METRICS = ["MOTA", "MOTP", "ID_consistency", "T_SR", "T_mIoU"]


def print_scene_and_global_averages(rows):
    """
    rows: list of MetricsRow objects
    Prints:
        - average metrics per scene
        - global average across all scenes
    """

    scene_metrics = defaultdict(lambda: defaultdict(list))

    # collect metrics
    for r in rows:
        for m in RADAR_METRICS:
            val = getattr(r, m, None)
            if val is not None:
                scene_metrics[r.scene][m].append(val)

    print("\n================ PER SCENE AVERAGES ================\n")

    scene_avg = {}

    for scene in sorted(scene_metrics.keys()):
        print(f"Scene: {scene}")
        scene_avg[scene] = {}

        for m in RADAR_METRICS:
            arr = np.array(scene_metrics[scene][m], dtype=float)
            avg = float(np.nanmean(arr)) if len(arr) > 0 else float("nan")
            scene_avg[scene][m] = avg

            print(f"  {m:15s}: {avg:.4f}")

        print()

    print("\n================ GLOBAL AVERAGE (ALL JSONs) ================\n")

    for m in RADAR_METRICS:
        vals = []
        for scene in scene_avg:
            vals.extend(scene_metrics[scene][m])

        arr = np.array(vals, dtype=float)
        avg = float(np.nanmean(arr)) if len(arr) > 0 else float("nan")

        print(f"{m:15s}: {avg:.4f}")

    print()
```
